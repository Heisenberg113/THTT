```python
import math
g = lambda x : ( (math.e)**sin(x) ) / 2
from sympy import *
x = symbols ('x')
```

```python
#Tim xap xi ban dau
plot ( abs( g ( x ) . diff () ) , (x , 1.5 , 1.7) )
```

```python
#Tim nghiem gan dung va sai so sau 3 buoc lap
q = 0.73
x0 = 1.5
for _ in range (3) :
    x = g ( x0 )
    ss = q / (1 - q ) * abs( x - x0 )
    x0 = x
    print (x , ss )
```

```python
#Tim nghiem voi sai so cho truoc 10^n va cho biet so buoc lap da thuc hien
q = 0.73
x0 = 1.5
n = 0
print (n , x , ss )
while True :
    x = g ( x0 )
    ss = q / (1 - q ) * abs( x - x0 )
    x0 = x
    n += 1
    print (n , x , ss )
    if ss < 10** -4:
        break
```
